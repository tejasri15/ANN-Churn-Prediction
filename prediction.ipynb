```python
import tensorflow as tf
import numpy as np
import pandas as pd
import pickle
from tensorflow.keras.models import load_model
```

```python
#load trained model , 
model = load_model('model.h5')

#load the encoder and scaler
with open('ohe_geo.pkl','rb') as file:
    ohe_geo = pickle.load(file)

with open('label_encoder_gender.pkl','rb') as file:
    label_encoder_gender = pickle.load(file)
    
with open('scaler.pkl','rb') as file:
    scaler = pickle.load(file)
```

```python

input_data ={
        'CreditScore':600,
    'Geography':'France'
    ,'Gender':'male'
    ,'Age':40
    ,'Tenure':3
    ,'Balance':100000
    ,'NumOfProducts':2
    ,'HasCrCard':1
    ,'IsActiveMember':1
    ,'EstimatedSalary':500000
}

```

```python
#one hot encoded geography
geo_encoded = ohe_geo.transform([[input_data['Geography']]]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded,columns = ohe_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
#encode categorial value
from sklearn.preprocessing import LabelEncoder

label_encoder_gender = LabelEncoder()
input_df['Gender']= label_encoder_gender.fit_transform(input_df['Gender'])
input_df
```

```python
#concate the one hot encoded
input_df = pd.concat([input_df.drop('Geography',axis=1),geo_encoded_df],axis=1)
input_df
```

```python
##scaling the input data
input_scaled = scaler.transform(input_df)
input_scaled
```

```python
#predict the churn
prediction = model.predict(input_scaled)
prediction
```

```python
prediction_proba = prediction[0][0]
prediction_proba
```

```python
if(prediction_proba>0.5):
    print("client likes this")
else :
    print("Client not likes this")
```
